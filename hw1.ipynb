### 4. The table below provides a training data set containing six observations, three predictors, and one qualitative response variable. Suppose we wish to use this dataset to make a prediction for Y when   X1 = X2 = X3 = 0 using K-nearest neighbors.

| Obs  | $X_1$| $X_2$| $X_3$| Y    |
|------|------|------|------|------|
|   1  |  0   |  4   |  0   | Green|
|   2  |  2   |  0   |  1   | Red  |
|   3  |  0   |  1   |  3   | Red  |
|   4  |  -1  |  1   |  2   | Green|
|   5  |  -3  |  0   |  1   | Green|
|   6  |  2   |  0   |  1   | Red  |

#### (a) Compute the Euclidean distance between each observation and the test point X1 = X2 = X3 = 0.

$$EuclideanDistance = \sqrt{\sum_{i=1}^{n} (q_i - p_i)^{2}}$$
##### Equation from: https://www.youtube.com/watch?v=hl3bQySs8sM

```python
p = np.array((0, 0, 0))

# Observation 1
obs1 = np.array((0, 4, 0))
obs1Dist = np.linalg.norm(p-obs1)

# Observation 2
obs2 = np.array((2, 0, 1))
obs2Dist = np.linalg.norm(p-obs2)

# Observation 3
obs3 = np.array((0, 1, 3))
obs3Dist = np.linalg.norm(p-obs3)

# Observation 4
obs4 = np.array((-1, 1, 2))
obs4Dist = np.linalg.norm(p-obs4)

# Observation 5
obs5 = np.array((-3, 0, 1))
obs5Dist = np.linalg.norm(p-obs5)

# Observation 6
obs6 = np.array((2, 0, 1))
obs6Dist = np.linalg.norm(p-obs6)

df = pd.DataFrame([[1, obs1Dist], [2, obs2Dist], [3, obs3Dist], [4, obs4Dist], 
                   [5, obs5Dist], [6, obs6Dist]], 
                  columns=["Observation", "Euclidean Distance"])
display(df)
```

#### (b) What is our prediction with K = 1? Why?

```python
# 1 nearest neighbor
df.nsmallest(1, "Euclidean Distance")
```

Here we see that Observation 2 is the closest to (0, 0, 0) so it is the nearest neighbor with the smallest Euclidean Distance from the point in question. Observation 2 has a Y value of Red so therefore we would predict Red when K = 1.   

It is worth noting that Observation 6 is the same point as Observation 2 so either could be used with K = 1, but they would both predict Y to be Red since both have Y = Red.

#### (c) What is our prediction with K = 3? Why?

```python
# 3 nearest neighbors
df.nsmallest(3, "Euclidean Distance")
```

Observations 2, 6 and 4 are the closest to the point $X_1 = X_2 = X_3 = 0$ because they have the three smallest Euclidean Distances. Observations 2 and 6 both have Y = Red while observation 4 has Y = Green. For nearest neighbors we could average the K closest responses. If we did so in this case we would probably get brown from 2 parts red and 1 part green. However, for categorical output it is more logical to take the majority so the prediction would be Red.

#### (d) If the Bayes decision boundary in this problem is linear but the data is noisy, then would we expect the best value for K with be larger or smaller? Why?

If the data is noisy but the boundary is linear, having a larger value for K would probably lead to better results. If the data has a fair amount of variability than increasing the number of samples you take would increase the liklihood that the majority would be correct. However, as k grows the bias might increase as well but the variance should decrease with inverse of k. Since we know the boundary is linear, we can be confident that the majority of data in an area should be correct so therefore the best value of K would probably be larger as this contributes to a steadier boundary.
